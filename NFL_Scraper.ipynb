```python
#import libraries
import numpy as np
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
#list of seasons to download
seasons = [str(season) for season in range(2014, 2023)]
print(f'number of seasons = {len(seasons)}')

#list of teams
teams = ['crd', 'atl', 'rav', 'buf', 'car', 'chi', 'cin', 'cle', 'dal', 'den', 'det', 'gnb', 'htx', 'clt', 'jax', 'kan', 'sdg', 'ram', 'rai', 'mia', 'min', 'nwe', 'nor', 'nyg', 'nyj', 'phi', 'pit', 'sea', 'sfo', 'tam', 'oti', 'was']
print(f'number of teams = {len(teams)}')
```

## **Scraping Game Data **

```python
#scrape gamelong data

#import libraries
import random
import time

#create empty dataframe to append
nfl_df = pd.DataFrame()

#Iterate through the seasons
for season in seasons:
    #iterate through the teams
    for team in teams:
      #set the url
      url = 'https://www.pro-football-reference.com/teams/' + team + '/' + season + '/gamelog/'
      print(url)

      #Get the offensive stats and set into dataframe
      off_df = pd.read_html(url, header=1, attrs ={'id' : 'gamelog' + season})[0]

      #Get the defensive stats and set into dataframe
      def_df = pd.read_html(url, header=1, attrs ={'id' : 'gamelog_opp' + season})[0]

      #concatenate the two dataframes (offense and defense) (along columns, axis =1)
      team_df = pd.concat([off_df, def_df], axis=1)

      #insert the season as a new column
      team_df.insert(loc=0, column='Season', value=season)

      #insert the team as a new column
      team_df.insert(loc=2, column='Team', value=team.upper())

      #concatonate the team gamelog to the aggregate dataframe (along rows, axis=0)
      nfl_df = pd.concat([nfl_df, team_df], ignore_index=True)
      print('Just concatenated the team gamelog of ' + team + ' in ' + season)

      #Pause the program to abide by website rules
      time.sleep(random.randint(7,8))

#Display the aggregate df
print(nfl_df)
```

```python
#Save the downloaded data to a csv file
dataframe_file = '/content/drive/My Drive/Colab Notebooks/NFL Info/nfl_gamelog.csv'
nfl_df.to_csv(dataframe_file, index=False)
```

# **Scraping Vegas Lines**

```python
#scraping vegas lines data

#import libraries
import random
import time

#create empty DF to append
veg_df = pd.DataFrame()

#Iterate through the seasons
for season in seasons:
    #iterate through the teams
    for team in teams:
      #set the url
      url = 'https://www.pro-football-reference.com/teams/' + team + '/' + season + '_lines.htm'
      print(url)

      #Get the vegas lines from ('vegas_lines' table)
      lines_df = pd.read_html(url, header=0, attrs ={'id' : 'vegas_lines'})[0]

      #Insert season column
      lines_df.insert(loc=0, column='Season', value=season)

      #insert team column
      lines_df.insert(loc=2, column='Team', value=team.upper())

      #concat team lines df into the aggregate df (along rows, axis=0)
      veg_df = pd.concat([veg_df, lines_df], ignore_index=True)
      print('Just concatenated the team lines of ' + team + ' in ' + season)

      #Pause the program to abide by website rules
      time.sleep(random.randint(7,8))

print(veg_df)
```

```python
#Save downloaded vegas data to csv
veg_dataframe_file = '/content/drive/My Drive/Colab Notebooks/NFL Info/nfl_vegas_lines_2014-2022.csv'
veg_df.to_csv(veg_dataframe_file, index=False)
```

```python
#reload saved df
veg_df = pd.read_csv(veg_dataframe_file)
print(veg_df.shape)
```

```python
#Drop columns from veg_df dataframe
veg_df = veg_df.drop(veg_df.columns[6:], axis=1)
print(veg_df.info(verbose=True))
```

```python
#rename columns
column_names = {'G#':'G', 'Over/Under':'Total'}
veg_df = veg_df.rename(columns=column_names)
print(veg_df.info(verbose=True))
```

```python
#The vegas lines contain lines from the regular season *and* post seeason
#Drop rows where G (games) >= 17 from 2014  - 2020
#Drop rows where G (games) >= 17 from 2021 - 2022 (18 game seasons now)
veg_df = veg_df.query('(Season <= 2020 and G < 17) or (Season >= 2021 and G < 18)')
print(veg_df.shape)
```

```python
#Create 'Home' column based on 'Opp' column in the original df
veg_df['Home'] = veg_df['Opp'].apply(lambda x: 0 if x[0] == '@' else 1)
```

```python
#Remove @ symbol from Opp column
veg_df['Opp'] = veg_df['Opp'].apply(lambda x: x[1:] if x[0] == '@' else x)

abbr_dict = {'OAK':'RAI', 'LVR':'RAI', 'STL':'RAM', 'LAR':'RAM', 'SDG':'LAC',
                 'IND':'CLT', 'HOU':'HTX', 'BAL':'RAV', 'ARI':'CRD', 'TEN':'OTI',
                 'NWE': 'NE', 'KAN': 'KC', 'TAM': 'TB', 'NOR': 'NO', 'SFO': 'SF',
                 'GNB': 'GB',  'NYG': 'NYG', 'NYJ': 'NYJ', 'DET': 'DET', 'CHI': 'CHI',
                 'CAR': 'CAR', 'ATL': 'ATL', 'CIN': 'CIN', 'CLE': 'CLE', 'DAL': 'DAL',
                 'DEN': 'DEN', 'JAX': 'JAX', 'MIA': 'MIA', 'MIN': 'MIN', 'PHI': 'PHI',
                 'PIT': 'PIT', 'SEA': 'SEA', 'WAS': 'WAS', 'BUF': 'BUF'}
veg_df = veg_df.replace({'Opp': abbr_dict})
```
